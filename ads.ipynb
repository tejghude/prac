```python
# 1.	 Illustrate a bar graph indicating the population in Canada by years.
import plotly.express as px
data_canada = px.data.gapminder().query("country == 'Canada'")
fig = px.bar(data_canada, x='year', y='pop')
fig.show()
```

```python
# 2.	Use Grouped bar graphs to show subgroups that need to be visualized on the graph. The subgroups should be differentiated by distinct colours.
import pandas as pd
import matplotlib.pyplot as plt
plotdata = pd.DataFrame({
    "2018":[57,67,77,83],
    "2019":[68,73,80,79],
    "2020":[73,78,80,85]},
    index=["Django", "Gafur", "Tommy", "Ronnie"])

plotdata.plot(kind="bar",figsize=(10, 5))
plt.title("FIFA ratings")
plt.xlabel("Footballer")
plt.ylabel("Ratings")
```

```python
# 3.	Plot stacked bar graphs are used to show dataset subgroups.
import pandas as pd
import matplotlib.pyplot as plt
plotdata = pd.DataFrame({
    "2018":[57,67,77,83],
    "2019":[68,73,80,79],
    "2020":[73,78,80,85]},
    index=["Django", "Gafur", "Tommy", "Ronnie"])
plotdata.plot(kind='bar', stacked=True,figsize=(10, 7))
plt.title("FIFA ratings")
plt.xlabel("Footballer")
plt.ylabel("Ratings")
```

```python
# 4.	Use a line graph is used to visualize a trend in data over intervals of time.
import matplotlib.pyplot as plt
import numpy as np
  
  
l=list(range(2017,2023))
x = np.array(l)
y = np.array([2, 4, 3,1,2,5])
  
plt.plot(x, y)
plt.xlabel("X-axis")  # add X-axis label
plt.ylabel("Y-axis")  # add Y-axis label
plt.title("line graph")  # add title
plt.show()
```

```python
# 5.	Show the study of more than one variable over the same period using Multiple line graphs.
import matplotlib.pyplot as plt
import numpy as np

l=list(range(2017,2023))
x = np.array(l)
y = np.array([2, 4, 3,1,2,5])  
y1 = np.array([1, 2, 4,2,3,2])
  

plt.plot(x, y)
plt.plot(x, y1)
  
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```

```python
# 6.	Illustrate numerical proportion, it is divided into slices. In a pie chart, for every slice, each of its arc lengths is proportional to the amount it represents.
pop_df = pd.DataFrame(
    {
        "population":[30,100,20,60,80],
    },
    index=["UK","India","France","Russia","China"]
)
plt.pie(pop_df["population"],labels=pop_df.index,autopct=lambda x:"{:.1f}%".format(x))
plt.show()
```
